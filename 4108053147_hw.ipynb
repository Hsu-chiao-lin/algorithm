```python
import timeit
import matplotlib.pyplot as plt
```

## Recursive

```python
def Fibonacci(n):
   
    # Check if input is 0 then it will
    # print incorrect input
    if n < 0:
        print("Incorrect input")
 
    # Check if n is 0
    # then it will return 0
    elif n == 0:
        return 0
 
    # Check if n is 1,2
    # it will return 1
    elif n == 1 or n == 2:
        return 1
 
    else:
        return Fibonacci(n-1) + Fibonacci(n-2)
```

```python
result1=[]
size1=[]
for n in range(10,60,10):
    size1.append(n) 
    result1.append(timeit.timeit('Fibonacci(n)','from __main__ import Fibonacci,n',number=5))
```

```python
plt.plot(size1,result1,'ro-', linewidth=1, markersize=10)
plt.xlim(10,50)
plt.ylim([min(result1),max(result1)]) 
plt.title('Fibonacci number')
plt.xlabel("size of S ") 
plt.ylabel("execution time") 
plt.show()
```

## Dynamic Programming

```python
def fibonacci(n):
     
    # Taking 1st two fibonacci numbers as 0 and 1
    f = [0, 1] 
    for i in range(2, n+1):
        f.append(f[i-1] + f[i-2])
    return f[n]
```

```python
result2=[]
size2=[]
for n in range(10,110,10):
    size2.append(n) 
    result2.append(timeit.timeit('fibonacci(n)','from __main__ import fibonacci,n',number=50))
```
