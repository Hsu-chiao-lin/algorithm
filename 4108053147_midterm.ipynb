```python
import numpy as np
import timeit
import matplotlib.pyplot as plt
```

##  8.(a) Linear search

```python
def linear_search(S, x):
    for j in range(0, N):
        if S[j] == x:
            return j
    return -1
```

```python
S = np.random.randint(1,100,10)
N = len(S)
x = np.random.randint(1,100)
index = linear_search(S, x)
print(S,x)
if index == -1:
    print("Element not found")
else:
    print("Element found at index: ", index)
```

```python
timeit.timeit('linear_search(S,x)','from __main__ import linear_search,S,x',number=5)
```

```python
x = np.random.randint(1,1000)
result=[]
for s in range(10,1010,10):
    S = np.random.randint(1,1000,s)
    N = len(S)
    result.append(timeit.timeit('linear_search(S,x)','from __main__ import linear_search,S,x',number=5))
print(result[:5])
```

## 8.(b) Complexity Analysis
### Best Case: O(1)
### Worst Case: O(N) 
### Average Case: O(N)
### Auxiliary Space: O(1)

## 8.(c) Line chart

```python
x = np.random.randint(100,10000)
result=[]
size=[]
for s in range(100,10010,10):
    size.append(s) 
    S = np.random.randint(100,10000,s) 
    N = len(S)
    result.append(timeit.timeit('linear_search(S,x)','from __main__ import linear_search,S,x',number=5))
print(result[:5])
```

## 8.(a) Binary search

```python
def binary_search(S, x):
    #
    low = 0
    upper = N - 1
    while low <= upper:
        mid = int((low + upper) / 2)  #
        if S[mid] < x:    #+1
            low = mid + 1
        elif S[mid] > x:  #+1
            upper = mid - 1
        else:                    #
            return mid
    return -1
```

```python
S = np.random.randint(1,100,10)
S.sort() 
N = len(S)
x = np.random.randint(1,100)
index = binary_search(S, x)
print(S,x)
if index >= 0:
    print("Element found at index: " + str(index))
else:
    print("Element not found")
```

```python
timeit.timeit('binary_search(S,x)','from __main__ import binary_search,S,x',number=5)
```

## 8.(a) Fibonacci search

```python
def FibonacciGenerator(n):
    if n < 1:
        return 0
    elif n == 1:
        return 1
    else:
        return FibonacciGenerator(n - 1) + FibonacciGenerator(n - 2)
    
def FibonacciSearch(S, x):
    m = 0 
    while FibonacciGenerator(m) < N: # 
        m = m + 1 
    offset = -1    
    while (FibonacciGenerator(m) > 1):
        i = min( offset + FibonacciGenerator(m - 2) , N - 1)
        #print('Current Element : ',S[i])
        if (x > S[i]):
            m = m - 1
            offset = i
        elif (x < S[i]):
            m = m - 2
        else:
            return i        
    if(FibonacciGenerator(m - 1) and S[offset + 1] == x):
        return offset + 1
    return -1
```

```python
S = np.random.randint(1,100,10)
S.sort() 
N = len(S)
x = np.random.randint(1,100)
index = FibonacciSearch(S, x)
print(S,x)
if index >= 0:
    print("Element found at index: " + str(index))
else:
    print("Element not found")
```
